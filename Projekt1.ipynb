## 3) Feature engineering

Wyczytanie danych wraz z ich modyfikacją w celu wykrycia wartości brakujących.

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
data = pd.read_csv("Wyniki_pomiarów_dot._monitoringu_chemizmu_gleb_ornych_Polski_w_2020_roku.csv", na_values=" ")
data.head(20)
```

Sprawdzenie liczby brakujących danych w zależności od kolumny wraz z procentowym udziałem wartości brakujących do wszystkich w kolumnie.<br>

Możemy zauważyć, że całkowity brak pomiarów (100%) występuję dla kolumn z zawartością "Be (mg/kg)" oraz "benzo(g,h,i)perylen (µg/kg)", co może wynikać z braku odpowiednich narzędzi pomiarowych. W kolumnach "fluoren (µg/kg)", "dibenzo(a,h)antracen (µg/kg)", "Na (cmol/kg)", "antracen (µg/kg)", "naftalen (µg/kg)", "Hg (mg/kg)", "Cd (mg/kg)", "Li (mg/kg)", "benzo(b)fluoranten (µg/kg)", "benzo(a)fluoranten (µg/kg)", "fenantren (µg/kg)" znajdują się dane w których, brakujące dane są w zdecydowanej większości (69-99%). Większość tych danych pochodzi z grupy wielopierścieniowych węglowodorów aromatycznych, ich braki mogą być związane z trudnością pomiarową oraz drogą i trudno dostępną aparaturą pomiarową, przez co pomiary tych wartości mogły celowo zostać nieprzeprowadzone (MNAR) lub brak jedenj z nich bezpośrednio wpływa na brak drugiej. Zmienne takie jak "zasolenie (mg KCl/100g)" oraz "przewodność (mS/m)" mogą być wartościami brakującymi typu MAR, ponieważ brak jedej zmiennej może wpływać. Natomiast zmienne dla których ilość wartości prakujących wynosi poniżej 1% możemy zakwalifikować do MCAR.

```python
A = pd.concat([data.isna().sum(), data.isna().mean(), 100*data.isna().mean().round(4) ], axis=1, keys = ["Ilość brakujących wartości", "Stosunek brakujących do wszystkich", "Udział procentowy"])
A = A.sort_values("Stosunek brakujących do wszystkich", ascending=False)
A
```

Sprawdzenie ilości unikatowych etykiet dla zmiennych jakościowych.

```python
print("Liczebnosc etykiet zmiennych jakościowych:\n")
columns = ['Numer Punktu', 'Województwo', 'Gmina', 'Powiat']
for col in columns:
    print(f'{col}: {len(data[col].unique())}')
```

Sprawdzenie ilości unikatowych etykiet dla zmiennych ilościowych.

```python
print("Liczebnosc etykiet zmiennych ilościowych:\n")
C = ['Numer Punktu', 'Województwo', 'Gmina', 'Powiat']
for col in data.columns:
    if col not in C:
        print(f'{col}: {len(data[col].unique())}')
```

Funkcji train_test_split używamy do dzielenia zbioru danych na zbiór uczacy i testowy. Ustawiamy takie parametry jak np. test_size co konfiguruję wielkość zioru testowego oraz random_state, który to ustawia ziarno generatora liczb losowych.<br>
Wybrane zmienne nie sa przypadkowe, ponieważ wartość "Na (cmol/kg)" i "T (cmol/kg)" mają wpływ na wartość "przewodność (mS/m)".

```python
col_name = ['Na (cmol/kg)', 'T (cmol/kg)']
X_train, X_test, y_train, y_test = train_test_split(data[col_name], data['przewodność (mS/m)'], test_size=0.3, random_state=0)
```

W tym miejscu możemy zaobserwować, że zbiory zostały podzielone zgodnie z zadanymi kryteriami tj. wielkość zbioru uczącego stanowi ok. 30% wielkości całego zbioru. Oczywiście proporcja nie zawsze będzie wynosiła idealne 30% ze względu na ilość elementów w całym zbiorze, jednak zawsze, będzie to, liczba najbliższa idealnej.

```python
print("X_train: {} X_test: {} {:.2f}%\n"
      "y_train: {} y_test: {} {:.2f}%".format(X_train.shape, X_test.shape, 
                                              X_test.shape[0] / data[col_name].shape[0] * 100,
                                              y_train.shape, y_test.shape, 
                                              y_test.shape[0] / data[col_name].shape[0] * 100))
```

Zbadanie kardynaloności zmiennych.
Analizując poniższe wyniki, okazuje się, że dla zmienncyh "Na (cmol/kg)" oraz "T (cmol/kg)" w zbiorze testowym znajdują się wartości unikalne, których nie ma w zbiorze treningowym.

```python
for col in col_name:
    unique_test = [x for x in X_test[col].unique() if x not in X_train[col].unique()]
    print(col, len(unique_test))
```

Sprawdzenie ilości wartości brakujących, zarówno dla zbioru uczącego, jak i testowego.

```python
def print_nans(df):
    for col in df.columns:
        print(col, df[col].isna().sum())

print_nans(X_train)
print_nans(X_test)
```

Zastąpienie brakujących wartości w kolumnach przez średnią z wartości pozostałych, a następnie zstąpinie pozostałych brakujących danych przez zera.

```python
mean_values = data.mean()
data.fillna(mean_values, inplace=True)
data.fillna(0, inplace=True)
data
```

```python
after_dict = {"X_train": [], "X_test": []}
for col in X_train.columns:
    after_dict["X_train"].append(len(X_train[col].unique()))
    after_dict["X_test"].append(len(X_test[col].unique()))
after_df = pd.DataFrame(after_dict, index=[col_name])
after_df
```

```python
before_dict = {"X_train": [], "X_test": []}
for col in col_name:
        before_dict["X_train"].append(len(X_train[col].unique()))
        before_dict["X_test"].append(len(X_test[col].unique()))
before_df = pd.DataFrame(before_dict, index=[col_name])
before_df
```

## Obliczenie skośności i kurtozy

Jedną z miar asymetrii zbioru jest skośność (skewness). Wskaźnik asymetrii równy zeru wskazuje na symetrię rozkładu zmiennej, wartość dodatnia oznacza asymetrię prawostronną (rozkład ma dłuższy prawy "ogon"), a wartość ujemna - asymetrię lewostronną (rozkład ma dłuższy lewy "ogon"). Skośność może okazać się przydatna w identyfikacji nieliniowych relacji między zmiennymi.<br>

W statystyce mamy do czynienia również z innymi miarami w kontekście analizy struktur. Często używaną jest miara koncentracji (skupienia). Rozróżniamy dwa rodzaje koncentracji: koncentracja wokół średniej oraz koncentrację oznaczającą nierównomierny podział zjawiska. Najpopularniejszą miarą skupienia wyników wokół wartości średniej jest kurtoza (kurtosis), która porównuje badany rozkład empiryczny z rozkładem normalnym. Im większa kurtoza, tym bardziej wysmukła jest krzywa liczebności, a zatem większa koncentracja wartości zmiennej wokół średniej. Jeżeli kurtoza < 0 rozkład jest bardziej spłaszczony od normalnego, a jeżeli kurtoza > 0 to rozkład bardziej wysmukły od normalnego. Może okazać się przydatna do identyfikacji nietypowych lub ekstremalnych obserwacji w danych.

```python
from scipy.stats import skew, kurtosis
for col in data.columns[9:]:
    skewness = skew(data[col])
    kurt = kurtosis(data[col])
    print("Skośność", col, ":", skewness)
    print("Kurtoza:", col, ":", kurt, "\n")
```

# Obliczenie korelacji między zmiennymi


Korelacja odnosi się do statystycznego miernika siły i kierunku związku między dwiema zmiennymi. Może być używana w analizie danych i modelowaniu statystycznym, aby zrozumieć, jakie są wzajemne relacje pomiędzy zmiennymi oraz w procesie budowy modeli predykcjyjnych.
W zależności od wielkości korelacji możemy określić siłę związku między cechami:<br>
• corr = 0 - współzależność nie występuje, brak korelacji, zmienne nie są nieskorelowane,<br>
• 0 < |corr| < 0.3 - słaby stopień współzależności,<br>
• 0.3 <= |corr| < 0.5 - średni stopień współzależności,<br>
• 0.5 <= |corr| < 0.7 - znaczny stopień współzależności,<br>
• 0.7 <= |corr| < 0.9 - wysoki stopień współzależności,<br>
• |corr| >= 0.9 - bardzo wysoki stopień współzależności,<br>
• |corr| = 1 - Współzależność całkowita (ścisłość), tzn. zakeżność funkcyjna międzyanymi cehcami.<br>

Na wykresie poniżej możemy zobaczyć iż, bardzo wysokim stopniem korelacji cechują się pary zmiennych takie jak "próchnica (%)" oraz "Corg (%)" czy "przewodność (mS/m)" oraz "zasolenie (mg KCI/100g)". Z kolei średnim stopniem współzależności objawiają się zmienne "T (cmol/kg)" oraz "Mg (mg 100g)" czy "pH H2O" oraz "Mg (mg 100g)". Większość zmienny jest skorelowana w stopniu słabym.
